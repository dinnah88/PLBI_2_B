```python
data_marketing['Monetary'] = data_marketing[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts']].sum(axis=1)
data_marketing
```

```python
## membuat rating recency dimulai dari bins untuk rating 0-14 < 14-30 < 30-45 < 45-75 < 75-100
data_marketing['Recency_Rating'] = pd.cut(x=data_marketing['Recency'], 
                                   bins=[0, 14, 30, 45, 75, 99], 
                                   labels=['5', '4', '3', '2', '1'], ## Label harus menerima tipe data str
                                   include_lowest=True, ## supaya nilai terendah (0) ikut masuk bin 
                                   )

## membuat rating monetary dimulai dari bins untuk rating (<100) < 1000-499 < 500-999 < 1000-2000 < (>2000)
data_marketing['Monetary_Rating'] = pd.cut(x=data_marketing['Monetary'], 
                                    bins=[0, 100, 500, 1000, 2001, np.inf], 
                                    labels=['1', '2', '3', '4', '5'], ## Label harus menerima tipe data str
                                    right = False)

## membuat rating frequency dimulai dari bins untuk rating (<100) < 1000-499 < 500-999 < 1000-2000 < (>2000)
data_marketing['Frequency_Rating'] = pd.cut(x=data_marketing['NumWebPurchases'], 
                                     bins=[0, 5, 10, 15, 21, np.inf], 
                                     labels=['1', '2', '3', '4', '5'], ## Label harus menerima tipe data str
                                     right = False)

data_marketing
```

```python
data_marketing['RFM_Rating'] = data_marketing[['Recency_Rating', 'Frequency_Rating', 'Monetary_Rating']].astype(int).mean(axis=1).apply(round)
data_marketing
```

```python
## Cutomer Labelling menggunakan regex 
segmentasi_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cannot_loose_them',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

data_marketing['Customer_Label'] =(data_marketing['Recency_Rating'].astype(str) + data_marketing['Frequency_Rating'].astype(str)).replace(segmentasi_map, regex=True)
data_marketing
```

```python
import pandas as pd
import numpy as np 

from statsmodels.stats.proportion import power_proportions_2indep, test_proportions_2indep
from statsmodels.stats.weightstats import ttest_ind
from statsmodels.stats.power import tt_ind_solve_power
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = df.copy().drop(columns=['user_id', 'landing_page'])
data
```

#### Ekstrak tanggal dari timestamp

```python
data['date'] = pd.to_datetime(data['timestamp']).dt.date
data.sort_values(by='timestamp')
```

#### Jumlah Visit, Converted yang digroup by date

```python
ab_test = data[['group', 'converted', 'date']]
ab_test
```

```python
# Mengambil value dari control_visit, treatment_visit, control_converted, treatment_converted 
value = ab_test.groupby(['date', 'group']).agg(['count', 'sum']).unstack().reset_index().values
value
```

```python
ab_test = pd.DataFrame(value, columns=['date_ab_test', 'control_visit', 'treatment_visit', 'control_converted', 'treatment_converted'])
ab_test
```

```python
ab_test = ab_test.set_index('date_ab_test').cumsum() #cummulative sum
ab_test
```

```python
ab_test['control_converted_rate'] = ab_test['control_converted'] / ab_test['control_visit'] 
ab_test['treatment_converted_rate'] = ab_test['treatment_converted'] / ab_test['treatment_visit'] 
ab_test
```

```python
def p_value(ab_test):

  result = test_proportions_2indep(count1=ab_test["treatment_converted"], 
                                   nobs1=ab_test["treatment_visit"], 
                                   count2=ab_test["control_converted"], 
                                   nobs2=ab_test["control_visit"],
                                   alternative="two-sided")
  return result[1]
```

```python
def power(ab_test):

  test_split = ab_test['treatment_visit'] / (ab_test['treatment_visit'] + ab_test['control_visit'])
  relative_effect = (ab_test['treatment_converted_rate'] - ab_test['control_converted_rate'])/ab_test['control_converted_rate']  

  power_result = power_proportions_2indep(diff=ab_test['control_converted_rate'] * relative_effect,
                                               prop2=ab_test['control_converted_rate'], 
                                               nobs1=ab_test['treatment_visit'],
                                               ratio=(1-test_split)/test_split,
                                               alpha=0.05,
                                               alternative='two-sided',
                                               return_results=False)
  return power_result
```

```python
ab_test['p_value'] = ab_test.apply(p_value, axis=1)
ab_test['power'] = ab_test.apply(power, axis=1)
ab_test
```

```python
ab_test.to_excel('ab_test_result.xlsx')
```

## Repeat Order

```python
customer = sales_performance.groupby('customer_unique_id').nunique()[['customer_id', 'order_purchase_timestamp']]
customer
```

```python
customer[(customer['customer_id'] > 1) & (customer['order_purchase_timestamp'] > 1)].sort_values(by='customer_id')
```

```python
customer = customer.reset_index()
```

```python
def status_repeat(df):
  if (df['customer_id'] > 1) and (df['order_purchase_timestamp'] > 1):
    return 'yes'
  return 'no'
```

```python
customer['Repeat_Order'] = customer.apply(status_repeat, axis=1)
customer
```

```python
customer.columns = ['customer_unique_id', 'banyak_customer_id', 'banyak_order_purchase_timestamp','Repeat_Order']
customer
```

```python
sales_marketing_repeat_order = pd.merge(sales_marketing, customer,
                               how='left',
                               on='customer_unique_id')
sales_marketing_repeat_order
```
